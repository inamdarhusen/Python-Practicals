# Groupby Function

- Instead of calculating average salaries separately , we can use pandas groupby method.


```python
#Grouby aggregate method
```

```python
df.groupby('rank')['salary'].mean()
```

```python
#min,max,mean,median
```

```python
df.groupby('rank')['salary'].agg({'min','max','mean','median'})
```
